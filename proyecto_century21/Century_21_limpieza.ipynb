Se revisan las filas que poseen la medida de terrno en HectÃ¡reas, se eliminan las filas sin info concluyente y se reemplaza los valores donde las hectÃ¡reas eran metros cuadrados

```python
df.drop(index=[22285,10822,20924,20745,21778,21759,21782],inplace=True)
df.loc[[17810, 20691, 20767, 20789], 'm2CSort'] = df.loc[[17810, 20691, 20767, 20789], 'm2C']
df.loc[[19503,10975,16240,21777], 'm2CSort'] = [450, 35.27, 0,0]
```

La informaciÃ³n para pies cuadrados calza correctamente

```python
filtro = (df['m2TTxt']=='Pies Cuadrados Terreno') & (df['m2CSort']>0) 
df[filtro][['tipoPropiedadTrans', 'm2TTxt', 'm2T', 'm2TSort', 'm2C', 'm2CSort']].sort_values(by='m2CSort', ascending=False)
```

Se transforma el campo de precio a uno vÃ¡lido

```python
df['precioSecundarioFormat'] = df['precioSecundarioFormat'].str.replace('$', '', regex=False)\
                                                           .str.replace(',', '', regex=False)\
                                                           .str.replace(' MXN', '', regex=False)
df['precioSecundarioFormat'] = df['precioSecundarioFormat'].astype(float)
df.head()
```

Se limpia la columna de cuota mantenimiento, dejando solo los valores numericos

```python
df['cuotaMantenimiento_limpia'] = df['cuotaMantenimiento'].str.extract('(\d+)', expand=False)
df['cuotaMantenimiento_limpia'] = pd.to_numeric(df['cuotaMantenimiento_limpia'], errors='coerce') # si existe algun valor incorrecto lo dejarÃ¡ como nulo
```

Se rellena con 0 todos los valores nulos

```python
df['cuotaMantenimiento_limpia'] = df['cuotaMantenimiento_limpia'].fillna(0)
```

Se suma el precio con la cuota de mantenimiento

```python
df['precio_con_mantenimiento'] = df['precioSecundarioFormat'] + df['cuotaMantenimiento_limpia']
df.head()
```

Se eliminan las columnas que no se necesitan

```python
df.drop(columns=['diasModificacionTxt','diasModificacionTxt','fotos', 'logoOficina', 'asesorFotoMostrarEnInternet', 'mostrarMonedaSecundaria','monedaSecundaria','precios', 'whatsapp',
                 'metaTags', 'etiquetas', 'asesorThumbnail','idAsesor', 'idAfiliados', 'idOriginal', 'id',
                 'idPropiedadesOS','idPropiedadesDB', 'idAfiliadosOS', 'fotos360', 'conVideo', 'conMapa', 'status', 'statusTrans', 'enInternet',
                 'idAfiliadosDB','idAsesoresOS', 'idAsesoresDB', 'mascotasTxtShort', 'albercaTxt', 'albercaTxtShort', 'recorridoVirtual',
                 'idAsesoresOriginalOS','idAsesoresOriginalDB', 'fechaAlta', 'banosTxtShort', 'estacionamientosTxt', 'estacionamientosTxtShort', 'mascotasTxt',
                 'fechaModificacion', 'ocultarCalleInternet', 'recamarasTxt', 'recamarasTxtShort', 'banosTxt',
                 'ocultarNumeroInternet', 'coloniaWeb', 'unidadDeMedida','municipioWeb', 'm2CTxt', 'estadoWeb', 'paisWeb',
                 'tipoOperacionTxt', 'precioMapa','tipoOperacionTrans', 'precio', 'moneda', 'ocultarPrecioInternet', 'pais', 'calle',
                 'tipoPropiedadEnTipoOperacion', 'tipoPropiedad','m2CFormat', 'm2TTxt', 'm2TFormat','m2T', 'm2C', 'precioFormat', 'telefono', 'email', 'asesorNombre', 'nombreAfiliado'
                 ], axis=1, inplace=True)
df.head()
```

Se separan los df de casa, departamento y terreno para trabajar por separado en cada uno de ellos

```python
df_casa = df[df['tipoPropiedadTrans']=='casa'].reset_index(drop=True)
df_dpto = df[df['tipoPropiedadTrans']=='departamento'].reset_index(drop=True)
df_terreno = df[df['tipoPropiedadTrans']=='terreno'].reset_index(drop=True)
```

Se calcula la moda de recamaras y baÃ±os por cada en cada estado

```python
# calcula la moda agrupando por estado de recamaras y baÃ±os, en caso de que haya mas de una moda se elige la primera
moda_recamaras = df_casa.groupby('estado')['recamaras'].agg(lambda x: pd.Series.mode(x)[0]) 
moda_baÃ±os = df_casa.groupby('estado')['banos'].agg(lambda x: pd.Series.mode(x)[0])

# Combinar los resultados en un solo DataFrame
modas = pd.concat([moda_recamaras.rename('moda_recamaras'), moda_baÃ±os.rename('moda_baÃ±os')], axis=1)

print(modas)
```

Se utiliza la tabla de modas por estado para rellenar los nulos de baÃ±os y recamaras en los registros de casas

```python
df_casa = pd.merge(df_casa, modas, on='estado', how='left')

# columna para identifcar columnas con nulos
df_casa.loc[df_casa['recamaras'].isna(),'recamaras_na'] = 1
df_casa.loc[df_casa['banos'].isna(),'banos_na'] = 1

# se rellenan los nulos con el promedio del estado
df_casa['recamaras'] = df_casa['recamaras'].fillna(df_casa['moda_recamaras']) 
df_casa['banos'] = df_casa['banos'].fillna(df_casa['moda_baÃ±os'])

# se eliminan las columnas de moda 
df_casa=df_casa.drop(columns=['moda_recamaras', 'moda_baÃ±os'])
df_casa
```

```python
df_casa[df_casa['recamaras_na'].notna()]
```

Se usa el rango intercuartÃ­lico para el tratamiento de outliers

```python
Q1_casa = df_casa['precioSecundarioFormat'].quantile(0.25)
Q3_casa = df_casa['precioSecundarioFormat'].quantile(0.75)
IQR_casa = Q3_casa - Q1_casa

lower_limit_casa = Q1_casa - 1.5 * IQR_casa
upper_limit_casa = Q3_casa + 1.5 * IQR_casa

outliers_casa = df_casa[(df_casa['precioSecundarioFormat'] < lower_limit_casa) | (df_casa['precioSecundarioFormat'] > upper_limit_casa)]
largo_outliers_casa = len(outliers_casa)
print(f'el lower limit de casa es {lower_limit_casa}')
print(f'el upper limit de casa es {upper_limit_casa}')
print(f'La cantidadd de outliers para el df de casas es {largo_outliers_casa}')
```

Se imputan los outliers con el valor del upper limit, no se usa el lower limit en la siguiente condiciÃ³n ya que es un valor negativo

```python
df_casa['precio_imputado'] = np.where(
    (df_casa['precioSecundarioFormat'] > upper_limit_casa),
    upper_limit_casa, 
    df_casa['precioSecundarioFormat']  
)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.boxplot(df_casa['precioSecundarioFormat'])
plt.title('Con Outliers')

plt.subplot(1, 2, 2)
plt.boxplot(df_casa['precio_imputado'])
plt.title('Outliers imputados')

plt.show()
```
